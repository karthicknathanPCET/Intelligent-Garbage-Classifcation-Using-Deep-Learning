```python
import pandas
```

```python
df = pandas.read_csv("House Price India.csv")
```

```python
df
```

```python
df.isna().any()
```

```python
df.shape
```

```python
X=pandas.DataFrame(columns=['longitude','latitude','living_area','lot_area','number_of_bedrooms','number_of_bathrooms','number_of_floors','waterfront_present','number_of_views','condition_of_the_house','Area_of_the_house','Area_of_the_basement','Built_year','Number_of_schools_nearby','Distanca_from_the_airport','Price'],data=df)
y=pandas.DataFrame(columns=['grade_of_the_house','Renovation','living_area_renov','lot_area_renov',],data=df)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
X_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

#Input Layer
model.add(Dense(X.shape[1], activation='relu', input_dim = X.shape[1]))

#Hidden Layer
model.add(Dense(512,kernel_initializer='normal', activation='relu'))
model.add(Dense(512,kernel_initializer='normal', activation='relu'))
model.add(Dense(256,kernel_initializer='normal', activation='relu'))
model.add(Dense(128,kernel_initializer='normal', activation='relu'))
model.add(Dense(64,kernel_initializer='normal', activation='relu'))
model.add(Dense(32,kernel_initializer='normal', activation='relu'))
#Output Layer
model.add(Dense(1,kernel_initializer='normal', activation = 'relu'))
```

```python
X.shape[1]
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
model.summary()
```

```python
history = model.fit(X_train, y_train.to_numpy(), batch_size = 10, epochs = 10, verbose = 1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.evaluate(X_test, y_test)
```
